```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Paths to your images
path_to_grid_search_images_dir = 'images/grid_search'
grid_search_images= [os.path.join(path_to_grid_search_images_dir, img) for img in os.listdir(path_to_grid_search_images_dir) if img.endswith('.png')]

path_to_results_images_dir = 'images/results'
results_images = [os.path.join(path_to_results_images_dir, img) for img in os.listdir(path_to_results_images_dir) if img.endswith('.png')]

#
```

```python
# Number of columns and rows
cols = 5
rows = 3

# Create a figure
fig, axes = plt.subplots(rows, cols, figsize=(15, 9))

# Plot grid search images in the first column
for i in range(len(grid_search_images)):
    img = mpimg.imread(grid_search_images[i])
    axes[i, 0].imshow(img)
    axes[i, 0].axis('off')

# Plot results images in the remaining columns
for i in range(len(results_images)):
    row = i // (cols - 1)
    col = i % (cols - 1) + 1
    img = mpimg.imread(results_images[i])
    axes[row, col].imshow(img)
    axes[row, col].axis('off')

# Remove axes for any remaining empty subplots
for r in range(rows):
    for c in range(cols):
        if (r == 2 and c == 0) or (r == 2 and c > len(results_images) % (cols - 1)):
            axes[r, c].axis('off')

# Add labels below the grid search images and results images
fig.text(0.2, 0.05, 'Grid Search Images', ha='center', fontsize=14, weight='bold')
fig.text(0.6, 0.05, 'Evaluation Images', ha='center', fontsize=14, weight='bold')

# Draw a dashed line between the first and second columns
line = plt.Line2D([0.2775, 0.2775], [0.1, 0.9], color='gray', linestyle='--', linewidth=2, transform=fig.transFigure)
fig.add_artist(line)

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.05, hspace=0.1)
plt.savefig('latex/appendix/dataset.png', dpi=300, bbox_inches='tight')
plt.show()
```
