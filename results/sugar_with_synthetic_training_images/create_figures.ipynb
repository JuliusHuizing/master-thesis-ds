# SuGar With Synthetic Training Images

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the directory containing the images
image_dir = 'dg_for_sugar/colmap/images'
save_path = '/Users/julius/repos/sisa3d/master-thesis-ds/latex/appendix/sugar_synthetic_training_images.png'

# List all image files in the directory
image_files = sorted([f for f in os.listdir(image_dir) if f.endswith('.png')])

# Select every 10th image
selected_images = image_files[::10]

# Create a figure and axes
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

# Plot the selected images
for ax, image_file in zip(axes.flat, selected_images):
    # Read the image
    img = mpimg.imread(os.path.join(image_dir, image_file))
    # Display the image
    ax.imshow(img)
    # Remove the axis
    ax.axis('off')
    # Set the title
    # ax.set_title(image_file)

# Adjust the layout
plt.tight_layout()
# Save the figure
plt.savefig(save_path)
plt.show()
```
